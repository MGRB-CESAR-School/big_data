```python
import dask.dataframe as dd
from dask.distributed import Client, LocalCluster
```

```python
cluster = LocalCluster()
client = Client(cluster)
client
```

```python
BASE_DATA_PATH = "../../data/"
BITCOIN_INPUT_PATH = BASE_DATA_PATH + "bitcoin/*/*.csv"
BITCOIN_OUTPUT_PATH = BASE_DATA_PATH + "data/bigfile.csv"
```

```python
%%time

df = dd.read_csv(BITCOIN_INPUT_PATH)
df
```

```python
%%time

df.to_csv(BITCOIN_OUTPUT_PATH, single_file=True, header=False, index=False)
```

```python
client.close()
cluster.close()
```
